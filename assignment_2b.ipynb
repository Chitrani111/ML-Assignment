```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

import matplotlib.pyplot as plt
from sklearn import linear_model
df = pd.read_csv("final_semester_marks.csv")
%matplotlib inline
```

```python
df
```

```python
# by code

# when X as independent variables (regression of Y on X)

X = df[['ML']]  # Features
Y = df['HRI']    # Target

# Perform linear regression
model = LinearRegression()
print(model.fit(X, Y))

# Print the slope and intercept
#print(f'Slope (Coefficient): {model.coef_[0]}')
#print(f'Intercept: {model.intercept_}')

# GRAPH
max_x = np.max(X)
min_x = np.min(X)
x = np.linspace(min_x, max_x, 1000)
W1 = model.coef_[0]
W0 = model.intercept_
y = W0 + W1 * x
plt.plot(x, y, color='orange', label='Regression line')
plt.xlabel('Marks obtained in ML')
plt.ylabel('Marks obtained in HRI')
plt.scatter(df.ML,df.HRI, color='purple', marker='*')
plt.legend()
plt.show()

# Function to predict unknown values of X
def predict_unknown_values(unknown_values):
    unknown_values = np.array(unknown_values).reshape(-1, 1)
    predictions = model.predict(unknown_values)
    return predictions

# Ask user for unknown X values
unknown_values = input("Enter unknown values of X, separated by commas: ").split(',')
unknown_values = [float(value) for value in unknown_values]

# Predict Y for unknown values of X
predictions = predict_unknown_values(unknown_values)
print(f'Predicted Y values: {predictions}')

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Predict on the test set
Y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(Y_test, Y_pred)
print(f'Mean Squared Error: {mse:.6f}')
```

```python
# using formula

# when X as independent variables (regression of Y on X)
%matplotlib inline
X = df['ML'].values
Y = df['HRI'].values
A = 0
B = 0
C = 0
D = 0
n = len(X)
for i in range(n):
    A += X[i]
    B += Y[i]
    C += X[i] * X[i]
    D += X[i] * Y[i]
W1 = (A*B - n*D) / (A*A -C*n)
W0 = (B*C - A*D) / (C*n - A*A)

max_x = np.max(X)
min_x = np.min(X)
x = np.linspace(min_x, max_x, 1000)
y = W0 + W1 * x
plt.plot(x, y, color='orange', label='Regression line')
plt.xlabel('Marks obtained in ML')
plt.ylabel('Marks obtained in HRI')
plt.scatter(df.ML,df.HRI, color='purple', marker='*')
plt.legend()
plt.show()

pred_hri = W0 + W1 * 96
print("\n If a student receives 96 in ML expected HRI is: ", pred_hri)

import math
error = 0
for i in range(n):
    error += ( abs( W1*X[i] - Y[i] + W0 ) / (math.sqrt( W1*W1 + 1*1)) )
avg_error = error / n
print("\n average predicted error is ", avg_error)

err = 0
for i in range(n):
    err += ((W0 + W1 * X[i]) - Y[i])**2
mse = err / (2*n)
print("\n The mean square error is ", mse)
```
